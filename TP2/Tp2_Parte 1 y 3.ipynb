```python
def reemplazar_valores(df, columnas):
    reemplazos = {
        "Sí": 1,
        "SÍ": 1,
        "No": 2,
        "NO": 2,
        "Ns./Nr.": 99
    }
    df[columnas] = df[columnas].replace(reemplazos)
    return df

# Lista completa de columnas
columnas_a_reemplazar = ["pp02c6", "pp02c7", "pp02c8", "pp02h", "pp02i", "pp03g", "pp03i", "pp03j", "pp05e",
    "pp06a", "pp06h", "pp07f1", "pp07f2", "pp07f3", "pp07f4", "pp07f5", "pp07g1",
    "pp07g2", "pp07g3", "pp07g4", "pp07h", "pp07i", "pp10c", "pp10d", "pp11n", "pp11p",
    "pp11q", "pp11r", "pp11s", "pp11t",
    "h15", "ch09", "ch13", "pp02c1", "pp02c2", "pp02c3", "pp02c4", "pp02c5"
]

# Aplicar reemplazo en df_2004
df_2004 = reemplazar_valores(df_2004, columnas_a_reemplazar)
```

```python
# Contar valores faltantes por año
faltantes_2004 = df_2004.isna().sum().to_frame(name="Faltantes_2004")
faltantes_2024 = df_2024.isna().sum().to_frame(name="Faltantes_2024")
```

```python
# Definir los valores faltantes (0, 99, y NaN)
valores_faltantes = [0, 99]

# Definir las variables de interés
variables_interes = [
    "CH04",  # Sexo
    "CH06",  # Edad
    "ESTADO",  # Condición de actividad
    "CAT_OCUP", # Categoría ocupacional
    "AGLOMERADO", # Código de Aglomerado
    "NIVEL_ED",  # Nivel educativo
    "P47T",  # Ingreso total individual
    "P21",  # MONTO DE INGRESO DE LA OCUPACIÓN PRINCIPAL
    "CH03",  # Relación de parentesco
    "CH07",  # Estado civil
    "CH08",  # ¿Tiene algún tipo de cobertura médica por la que paga o le descuentan?
    "CH10",  # ¿Asiste o asistió a algún establecimiento educativo (colegio, escuela, universidad)?
    "CH11",  # Ese establecimiento es público, privado o Ns./Nr.?
    "CH12",  # Nivel cursado
    "CH13",  # ¿Lo finalizó ese nivel?
]

# Filtrar los datos para cada año (2004 y 2024) usando la columna 'ano4'
df_2004 = df_unido[df_unido['ANO4'] == 2004]
df_2024 = df_unido[df_unido['ANO4'] == 2024]

# Contar los valores faltantes en 2004 para las variables de interés
faltantes_2004 = df_2004[variables_interes].isin(valores_faltantes).sum() + df_2004[variables_interes].isna().sum()

# Contar los valores faltantes en 2024 para las variables de interés
faltantes_2024 = df_2024[variables_interes].isin(valores_faltantes).sum() + df_2024[variables_interes].isna().sum()

# Convertir las series a DataFrame para mostrar los resultados
faltantes_2004_df = faltantes_2004.to_frame(name="Faltantes_2004")
faltantes_2024_df = faltantes_2024.to_frame(name="Faltantes_2024")

# Mostrar los resultados
print("Faltantes 2004:")
print(faltantes_2004_df)
print("\nFaltantes 2024:")
print(faltantes_2024_df)
```

```python
# PARTE 3: Punto 5 y 6 --- Decidimos ya definir la base de datos "respondieron" para tener los datos ordenados desde el principio.
respondieron = df_unido[(df_unido["ESTADO"].notna()) & (df_unido["ESTADO"] != 0)]
```

```python
# Ahora sí, contamos la cantidad de personas desocupadas (ESTADO = 2)
desocupados_total = respondieron[respondieron["ESTADO"] == 2]
print("Cantidad de personas desocupadas:", len(desocupados_total))
```

```python
# Ahora las inactivas
inactivos_total = respondieron[respondieron["ESTADO"] == 3]
print("Cantidad de personas inactivas:", len(inactivos_total))
```

```python
# Calculamos el promedio del ingreso per cápita familiar (IPCF) según condición de actividad (ESTADO) y redondeamos a dos decimales para mejor lectura
# ESTADO = 1 → Ocupado
# ESTADO = 2 → Desocupado
# ESTADO = 3 → Inactivo

ipcf_promedio = respondieron.groupby("ESTADO")["IPCF"].mean().round(2)

print("Promedio de ingreso per cápita familiar por estado:")
print(ipcf_promedio)
```

```python
# Creamos la base de quienes NO respondieron su condición de actividad (ESTADO = 0 o nulo)
norespondieron = df_unido[(df_unido["ESTADO"].isna()) | (df_unido["ESTADO"] == 0)]

# Mostramos cuántas personas hay en esa base
print("Cantidad de personas que no respondieron su condición de actividad:", len(norespondieron))
```

```python
# Punto 7: Creamos la columna PEA (Población Económicamente Activa)
respondieron.loc[:, "PEA"] = respondieron["ESTADO"].apply(lambda x: 1 if x in [1, 2] else 0)
```

```python
# Chequeamos que se creo la columna
respondieron[["ESTADO", "PEA"]].head()
```

```python
# Ahora graficamos
import matplotlib.pyplot as plt

# Agrupamos por año y por condición PEA
pea_por_año = respondieron.groupby(["AÑO", "PEA"]).size().unstack().fillna(0)

# Gráfico de barras apilado
pea_por_año.plot(kind='bar', stacked=True, figsize=(8,5))

plt.title("Composición de la PEA por año (2004 vs 2024)")
plt.xlabel("Año")
plt.ylabel("Cantidad de personas")
plt.legend(title="PEA", labels=["No PEA", "PEA"])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Convertimos la columna CH06 (edad) a valores numéricos.
# Esto es necesario porque algunos valores podrían venir como texto (strings),
# y no podríamos compararlos con números (ej: 15 <= x <= 65).
respondieron["CH06"] = pd.to_numeric(respondieron["CH06"], errors="coerce")
```

```python
# Punto 8: Creamos la nueva columna PET (Población en Edad para Trabajar) = 1 a quienes tienen entre 15 y 65 años inclusive (están en edad de trabajar) y 0 a quienes están por fuera de ese rango.
respondieron.loc[:, "PET"] = respondieron["CH06"].apply(lambda x: 1 if 15 <= x <= 65 else 0)
```

```python
# Punto 8: Gráficamos PET por año

import matplotlib.pyplot as plt

# Agrupamos por año y condición PET (1 = en edad, 0 = fuera de edad)
pet_por_año = respondieron.groupby(["AÑO", "PET"]).size().unstack().fillna(0)

# Gráfico de barras apilado
pet_por_año.plot(kind='bar', stacked=True, figsize=(8,5))

plt.title("Composición de la PET por año (2004 vs 2024)")
plt.xlabel("Año")
plt.ylabel("Cantidad de personas")
plt.legend(title="PET", labels=["Fuera de edad", "En edad de trabajar"])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Comparamos visualmente la composición de la PET y la PEA por año

import matplotlib.pyplot as plt

# Agrupamos por año y cada condición
pet_por_año = respondieron.groupby(["AÑO", "PET"]).size().unstack().fillna(0)
pea_por_año = respondieron.groupby(["AÑO", "PEA"]).size().unstack().fillna(0)

# Creamos figura con dos subplots uno al lado del otro
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de PET
pet_por_año.plot(kind='bar', stacked=True, ax=ax[0], color=["#1f77b4", "#ff7f0e"])
ax[0].set_title("Composición de PET por año")
ax[0].set_xlabel("Año")
ax[0].set_ylabel("Cantidad de personas")
ax[0].legend(title="PET", labels=["Fuera de edad", "En edad de trabajar"])
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico de PEA
pea_por_año.plot(kind='bar', stacked=True, ax=ax[1], color=["#1f77b4", "#ff7f0e"])
ax[1].set_title("Composición de PEA por año")
ax[1].set_xlabel("Año")
ax[1].set_ylabel("Cantidad de personas")
ax[1].legend(title="PEA", labels=["No PEA", "PEA"])
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Punto 9: Creamos la columna 'desocupado' en la base 'respondieron'. La persona está desocupada si ESTADO == 2.
respondieron.loc[:, "desocupado"] = respondieron["ESTADO"].apply(lambda x: 1 if x == 2 else 0)
```

```python
# Agrupamos por año y contamos la cantidad de personas desocupadas
desocupados_por_año = respondieron.groupby("AÑO")["desocupado"].sum()

print("Cantidad de personas desocupadas por año:")
print(desocupados_por_año)
```

```python
## Observación sobre la cantidad de personas desocupadas por año 

# Según los datos de la muestra relevada por la EPH para la región NOA, se observa que la cantidad de personas desocupadas en la muestra se redujo entre los años 2004 y 2024.

# En 2004, se registraron 597 personas desocupadas dentro del total de personas que respondieron su condición de actividad. En 2024, ese número desciende a 256.
```

```python
## Nos intereso hacer la siguiente conclusión final:
# Total de personas activas (PEA = 1) por año
pea_totales = respondieron[respondieron["PEA"] == 1].groupby("AÑO").size()

# Total de personas desocupadas por año
desocupados = respondieron[respondieron["desocupado"] == 1].groupby("AÑO").size()

# Calculamos ocupados = PEA - desocupados
ocupados = pea_totales - desocupados

# Tasa de ocupación dentro de la PEA
tasa_ocupacion = (ocupados / pea_totales * 100).round(2)

# Mostramos resultados
print("Tasa de ocupación dentro de la PEA (%):")
print(tasa_ocupacion)
```

```python
## CONCLUIÓN FINAL

# Según los cálculos realizados, se observa que la tasa de ocupación dentro de la Población Económicamente Activa (PEA) aumentó entre 2004 y 2024 en la región NOA, según los datos de la EPH.

# En 2004, el 84.82% de las personas activas estaban ocupadas.
# En 2024, ese porcentaje asciende al 94.34%.

# Esto implica que, si bien la cantidad total de personas activas no cambió drásticamente, una mayor proporción de ellas tenía trabajo en 2024. Es decir, dentro del grupo que buscaba insertarse en el mercado laboral, hubo una menor incidencia del desempleo en el segundo año analizado.

# Como se indicó anteriormente, estos resultados provienen de la muestra de la EPH y no deben extrapolarse a la población general sin considerar el uso de factores de expansión como `PONDERA`.
```
